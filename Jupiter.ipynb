```python
from transformers import pipeline
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
# Загружаем модель для русского языка
sentiment_pipeline = pipeline("sentiment-analysis",
                              model="blanchefort/rubert-base-cased-sentiment",
                              device=-1)  # device=-1 означает CPU (на телефоне работает)
print("Модель загружена!")
```

```python
def analyze_sentiment(text):
    """
    Принимает текст, возвращает словарь с меткой и уверенностью.
    """
    result = sentiment_pipeline(text)[0]
    # result = {'label': 'POSITIVE' или 'NEGATIVE' или 'NEUTRAL', 'score': 0.99}
    label = result['label']
    score = result['score']

    # Переводим метки на русский для красоты
    if label == 'POSITIVE':
        ru_label = 'Позитивный'
    elif label == 'NEGATIVE':
        ru_label = 'Негативный'
    else:
        ru_label = 'Нейтральный'

    return {
        'текст': text,
        'тональность': ru_label,
        'уверенность': round(score, 4)
    }
```

```python
test_text = "Этот фильм просто великолепен! Я в восторге."
print(analyze_sentiment(test_text))
```

```python
test_text2 = "Обслуживание ужасное, еда холодная, больше не приду."
print(analyze_sentiment(test_text2))
```

```python
examples = [
    "Я обожаю этот ресторан, всё очень вкусно!",
    "Сегодня на улице идёт дождь, но настроение нормальное.",
    "Этот товар сломался на второй день, полное разочарование.",
    "Купил новый телефон, пока доволен.",
    "Фильм не понравился, скучный и затянутый."
]

results = []
for ex in examples:
    results.append(analyze_sentiment(ex))

# Выводим таблицу
df = pd.DataFrame(results)
print(df)
```
